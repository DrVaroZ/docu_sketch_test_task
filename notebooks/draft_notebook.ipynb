```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_json('https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json')
```

```python
df.head(n=15)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.scatter(df['gt_corners'], df['rb_corners'])
plt.show()
```

```python
df.plot(x='name', y=['gt_corners', 'rb_corners'], kind='line', figsize=(20, 10))
```

```python
df.plot(x='gt_corners', y='rb_corners', kind='line', figsize=(10, 10))
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['gt_corners'], df['rb_corners'], c='crimson')

p1 = max(max(df['rb_corners']), max(df['gt_corners']))
p2 = min(min(df['rb_corners']), min(df['gt_corners']))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
accuracy_score(df['gt_corners'], df['rb_corners'])
```

as we can see, all predicted values are equal to true values.
so the model seems to be overfitting.
