```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def display(filename: str) -> None:
    """
    Function for reading json file and building plots

    :param filename: str - name of the file with plot image

    :return: None - shows image of the plot
    """

    full_path = 'D:\Python projects\docu_sketch_test_task\plots\\' + filename
    image = mpimg.imread(full_path)
    plt.figure(figsize=(20, 20))
    plt.imshow(image)
```

```python
path_to_plots = 'D:\Python projects\docu_sketch_test_task\plots'
plot_images = os.listdir(path_to_plots)
for file in plot_images:
    display(file)
```

Above we can see plots of dependency between true and predicted values - direct proportionality.
Also we can note that mean deviation linearly depends on floor_mean and ceiling_mean.
